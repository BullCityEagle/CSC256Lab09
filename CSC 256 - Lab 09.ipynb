```python
duckSearchResults = []
```

```python
for p in range(len((duckDump.json()['RelatedTopics']))):
    print('------------- ', p + 1 , ' -------------------------')
    print(duckDump.json()['RelatedTopics'][p]['Text'])
    print('')
    duckSearchResults.append(duckDump.json()['RelatedTopics'][p]['Text'])
```

```python
print(duckSearchResults)
```

```python
for result in duckSearchResults:
    if 'Lincoln' in result:
        print('16')
        break
```

```python
presidents = {1: 'Washington', 2: 'Adams', 3: 'Jefferson', 4: 'Madison', 5: 'Monroe', 6: 'McGlotten'}
```

```python
for p in range(len(presidents)):
    for result in duckSearchResults:
        if presidents[p+1] in result:
            print(presidents[p+1])
            break
```
